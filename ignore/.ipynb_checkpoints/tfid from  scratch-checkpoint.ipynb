```python
from __future__ import division
import math 
import string
import csv
import re
```

```python
def csv_reader(file, header=True):
    '''Takes in a csv file and returns a tokenized list of the data seperated as X_train, y_train and X_foreign'''
    X_train = []
    y_train = []
    train_coutn = 0

    X_foreign = []
    ext_count = 0
    with open(file, newline='') as csvfile:
        readCSV = csv.reader(csvfile)#, delimiter=',')
        for line in readCSV: #Iterate through the loop to read line by line
            if line[-1] == '':
                X_foreign.append(re.findall(r"[\w']+",line[0]))
                ext_count += 1
            else:
                X_train.append(' '.join(re.findall(r"[\w']+",line[0])))
                y_train.append(' '.join(re.findall(r"[\w']+",line[-1])))

                train_coutn += 1
                
        csvfile.close()
        
    if header:
        X_train.pop(0)
        y_train.pop(0)
        
    return X_train, y_train, X_foreign
```

```python
X_train, y_train, X_foreign = csv_reader('Data/Hotel_reviews_dataset.csv')
```

```python
corpus = sent_tokens(ohio)

for i in range(len(corpus )):
    corpus [i] = corpus [i].lower()
wordfreq = {}
for sentence in corpus:
    tokens = word_token(sentence)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
most_freq = sorted(wordfreq, key=wordfreq.get, reverse=True)[:200]
word_idf_values = {}
for token in most_freq:
    doc_containing_word = 0
    for document in corpus:
        if token in word_token(document):
            doc_containing_word += 1
    word_idf_values[token] = np.log(len(corpus)/(1 + doc_containing_word))
word_tf_values = {}
for token in most_freq:
    sent_tf_vector = []
    for document in corpus:
        doc_freq = 0
        for word in word_token(document):
            if token == word:
                  doc_freq += 1
        word_tf = doc_freq/len(word_token(document))
        sent_tf_vector.append(word_tf)
    word_tf_values[token] = sent_tf_vector
tfidf_values = []
for token in word_tf_values.keys():
    tfidf_sentences = []
    for tf_sentence in word_tf_values[token]:
        tf_idf_score = tf_sentence * word_idf_values[token]
        tfidf_sentences.append(tf_idf_score)
    tfidf_values.append(tfidf_sentences)
```

```python
tf_idf_model = np.asarray(tfidf_values)
print(tf_idf_model)
```
